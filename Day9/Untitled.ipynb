<h1>Mini Project: Simple Banking System with Exception Handling</h1>

```python
create a simple program simulate a bank account withdrawl process

1)ask user for the account balance and amount they wnat to withdraw
2)check for the invalid input
3)handle cases withdrawl more money then the available or trying to withdraw negative amount
4)handle where input is not a number
5)withdrawl amount is greater then the account balance
6)withdrawl amount is negative
```

```python
try:
    Balance = float(input("Enter your account balance: "))
except ValueError:
    print("Error: Invalid input! Please enter a valid number for the balance!")
else:
    try:
        withdrawl_amount = float(input("Enter the amount you want to withdraw: "))
    except ValueError:
        print("Error: Invalid input! Please enter a valid number for the withdrawl_amount!")
    else:
        if withdrawl_amount>Balance:
            print("Insufficient fund,you cannot withdraw more then your balance")
        elif withdrawl_amount<0:
            print("You cannot withdraw negative amount")
        else:
            Balance -= withdrawl_amount
            print(f"Transaction Successfull!your remaining balance is {Balance}")
        
    
```
