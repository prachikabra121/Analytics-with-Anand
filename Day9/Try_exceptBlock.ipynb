```python
try:
    num = int(input("Enter a number:"))
    result = 100/num
    print(result)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Please enter correct integer value")

else:
    print("NO exception occured")
    

finally:
    print("Alwys need to exceute this block")
```

```python
try:
    num = int(input("Enter a number:"))
    result = 100/num
    print(result)

except ZeroDivisionError as e:
    raise ZeroDivisionError("Cannot divide by zero") from e

except ValueError:
    print("Please enter correct integer value")

else:
    print("NO exception occured")
    

finally:
    print("Alwys need to exceute this block")
```
