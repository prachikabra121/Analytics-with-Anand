```python
l3=[]

for i in l:
    if type(i) in [list,set,tuple]:
        for j in i:
            l3.append(j)
    elif type(i) == dict:
        for k,v in i.items():
            l3.append(k)
            if type(v) in [list,set,tuple]:
                for h in v:
                    l3.append(h)

    else:
        l3.append(i)

print(l3)
```

```python
def toatl_bill(name,*prices):
    total = 0
    for i in prices:
        total += i

    return name,total
```

```python
toatl_bill("Anupam",12,10,30,40,50)
```

```python
def toatl_bill(name,*prices,**kwargs):
    total = 0
    for i in prices:
        total += i

    return name,total,kwargs
```

```python
toatl_bill("Anupam",12,10,30,40,50,Billing_address="abc",contact="123456789")
```

```python
toatl_bill("Anupam",Billing_address="abc",contact="123456789",12,10,30,40,50)
```

```python
#create a function to check the number is prime or not 


def prime(n):
    if n<=1:
        return False
    for i in range(2, n):
        if n%i==0:
            return False
    return True
```

```python
prime(9)
```

```python
prime(25)
```

```python
prime(13)
```

```python
n = int(input("Enter the number: "))
if n <= 1:
    print(False)
else:
    is_prime = True  # Flag variable
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break
    print(is_prime)
```
