```python
def fact(n):
    if n == 1:
        return 1
    else:
        return n*fact(n-1)
```

```python
fib(n) = fib(n-1)+fib(n-2)
fib(0) = 0
fib(1) = 1
```

```python
def fibb(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibb(n-1)+fibb(n-2)
```

```python
def sum_list(nums):
    if len(nums) ==0:
        return 0
    else:
        return nums[0]+sum_list(nums[1:])
        

```
