```python
old = "06/12/2025"
dt = datetime.strptime(old, "%d/%m/%Y")

new_format = dt.strftime("%Y-%m-%d")
print(new_format)
```

```python
| Format | Meaning          | Example  |
| ------ | ---------------- | -------- |
| `%d`   | Day (2 digits)   | 06       |
| `%m`   | Month (2 digits) | 12       |
| `%b`   | Month short name | Dec      |
| `%B`   | Month full name  | December |
| `%y`   | Year (2 digits)  | 25       |
| `%Y`   | Year (4 digits)  | 2025     |
| `%H`   | Hours (24-hour)  | 17       |
| `%I`   | Hours (12-hour)  | 05       |
| `%p`   | AM/PM            | PM       |
| `%M`   | Minutes          | 30       |
| `%S`   | Seconds          | 10       |
```
