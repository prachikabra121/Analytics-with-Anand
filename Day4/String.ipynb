```python
literal=input("enter literal")
if literal.isdigit():
    print("Integer")
elif literal.count('.')==1:
    print("float")
elif literal.isalpha():
    print("string")
elif literal.lower()in['true','false']:
    print("boolean")
elif literal.lower=="none":
    print("none")
else:
    print("other")
```
