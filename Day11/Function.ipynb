```python
l3=[]

for i in l:
    if type(i) in [list,set,tuple]:
        for j in i:
            l3.append(j)
    elif type(i) == dict:
        for k,v in i.items():
            l3.append(k)
            if type(v) in [list,set,tuple]:
                for h in v:
                    l3.append(h)

    else:
        l3.append(i)

print(l3)
```

```python
#
l1 = []
for i in l:
    if type(i) == int:
        l1.append(i)
    elif type(i) == tuple or type(i) == set or type(i)==list:
        for j in i:
            if type(j) == int:
                l1.append(j)
    elif type(i) == dict:
        for k,v in i.items():
            if type(v)==int:
                l1.append(v)
            elif type(v) in [list,tuple,set]:
                for h in v:
                    l1.append(h)
            elif type(k) == int:
                l1.append(k)
        

print(l1)
```

```python
listt=[]
for i in l:
    if type(i)==int:
        listt.append(i)
    elif type(i)==dict:
        for k,v in i.items():
            if type(k)==int:
                listt.append(k)
            elif type(v)==int:
                listt.append(v)
            elif type(v) in [list, tuple, set]:
                for j in v:
                    if type(j) == int:
                        listt.append(j)
    elif type(i) in [list, tuple, set]:
        for j in i:
            if type(j) == int:
                listt.append(j)
print(listt)
```

```python
#
l2 = []
for i in l:
    if type(i) == str:
        l2.append(i)
    elif type(i) == tuple or type(i) == set or type(i)==list:
        for j in i:
            if type(j) == str:
                l2.append(j)
    elif type(i) == dict:
        for k,v in i.items():
            if type(v)==str:
                l2.append(v)
            elif type(k) == str:
                l2.append(k)
        

print(l2)
```
