```python
# Open a file in read mode
file = open('Data.txt.txt', 'r')

```

```python
content = file.read()            # Read the entire content of the file
print(content)                   # Print content
   
```

```python
# Open a file in write mode (creates a new file if it doesn't exist)
file = open('Data.txt.txt', 'w')
```

```python
content = file.write("Everyday is beautiful')            # Read the entire content of the file
print(content)
```

```python
# Open a file in append mode (adds new content to the end)
file = open('example.txt', 'a')
```

```python
# Open a file in read-write mode
file = open('example.txt', 'r+')
```

```python
# Close the file after operation
file.close()
```
